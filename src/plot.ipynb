```python
period_dict = dict()
period_number = 1
is_attack_period = False

period_dict[period_number] = {
    "start_period": df.index[0],
    "class": "Normal"
}

for i in range(len(df)):
    if (not is_attack_period and df.loc[df.index[i]]["Normal/Attack"] == 'Attack'):
        is_attack_period = True
        period_number += 1
        period_dict[period_number] = {
            "start_period": df.index[i],
            "class": "Attack"
        }
        period_dict[period_number-1]["end_period"] = df.index[i-1]

    if (is_attack_period and df.loc[df.index[i]]["Normal/Attack"] == 'Normal'):
        is_attack_period = False
        period_number += 1
        period_dict[period_number] = {
            "start_period": df.index[i],
            "class": "Normal"
        }
        period_dict[period_number-1]["end_period"] = df.index[i-1]

period_dict[period_number]["end_period"] = df.index[len(df)-1]

print(period_dict)
```

```python
plt.figure(figsize=(200, 100))
plt.plot(detected_attacks.index, np.full((len(detected_attacks),), 1), marker="o", linestyle="", color="g")
# detected_attacks["FIT101"].plot(marker="o", linestyle="", color="g")
for i in range(1, period_number+1):
    if period_dict[i]["class"] == "Attack":
        plt.axvline(period_dict[i]["start_period"], color='r')
        plt.axvline(period_dict[i]["end_period"], color='b')

plt.show()
```
