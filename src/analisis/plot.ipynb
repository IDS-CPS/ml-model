```python
import pandas as pd
import joblib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../dataset/pompa-smooth-attack.csv")

df = df[["adc_level","adc_temp","adc_flow","adc_pressure_left","adc_pressure_right"]]
scaler = joblib.load("../../scaler/pit/uae-5.gz")

scaled = scaler.transform(df)
```

```python
scaled_df = pd.DataFrame(scaled, columns = ["adc_level","adc_temp","adc_flow","adc_pressure_left","adc_pressure_right"])
scaled_df = scaled_df[5:]
scaled_df.head()
```

```python
predict_df = pd.read_csv("../../dataset/pit/test-predict.csv")

predict_df.head()
```

```python
for column in ["adc_level","adc_temp","adc_flow","adc_pressure_left","adc_pressure_right"]:
    plt.figure(figsize=(100,100))

    ax = scaled_df[[column]].plot()
    predict_df[[column]].plot(ax=ax, color='red', linewidth=3)

    plt.show()
```
