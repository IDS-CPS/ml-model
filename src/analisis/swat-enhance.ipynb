```python
import pandas as pd
```

```python
df = pd.read_csv("../../dataset/swat-2015-data.csv", delimiter=";", decimal=",")
df = df[16000:]
df = df[::5]
```

```python
df.head()
```

```python
features_dropped = ["Timestamp", "Normal/Attack", "AIT201", "AIT202", "AIT203", "P201", "AIT401",
"AIT402", "AIT501", "AIT502", 'AIT503', "AIT504", "FIT503", "FIT504",
"PIT501", "PIT502", "PIT503", "P603"]

new_df = df.drop(columns=features_dropped)
new_df.head()
```

```python
for column in new_df.columns:
    new_df[f"{column}-past"] = new_df[column].shift(1)

new_df = new_df.reset_index(drop=True)
new_df = new_df[1:]
new_df.head()
```

```python
new_df.to_csv("../../dataset/swat-enhanced.csv", index=False)
```
