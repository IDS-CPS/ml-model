```python
import pandas as pd
```

```python
df = pd.read_csv("../../dataset/pompa-train-v3.csv")
```

```python
df.head()
```

```python
new_df = df
for column in df.columns:
    new_df[f"{column}-past"] = new_df[column].shift(1)

new_df = new_df.reset_index(drop=True)
new_df = df[1:]
new_df.head()
```

```python
new_df.to_csv("../../dataset/pompa-train-v3-enhanced.csv", index=False)
```
